```
import pandas as pd
```

```
original_matrix = pd.read_csv('/work/original_matrix.csv')
cosine_df = pd.read_csv('/work/cosine_games.csv', index_col=0)
svd_df = pd.read_csv('/work/svd_games.csv', index_col=0)
gower_distances_original = pd.read_csv('gower_distances_original.csv', index_col=0)
gower_distances_reduced = pd.read_csv('gower_distances_reduced.csv', index_col=0)
```

```
relavant_cols = ['Name_Year', 'YearPublished', 'mechanics', 'category', 'themes', 'MinPlayers', 'MaxPlayers']
```

# Game

```
game = 'Monopoly (1933)'
```

# Cosine Similarity

```
similars_cosine = cosine_df.loc[game].tolist()
similars_cosine
```

```
similars_cosine.insert(0,game)
```

```
original_matrix[original_matrix['Name_Year'].isin(similars_cosine)][relavant_cols].sort_values(by=['Name_Year'], key=lambda x: x.map({v: i for i, v in enumerate(similars_cosine)}))
```

# Singular Value Decomposition

```
similars_svd = svd_df.loc[game].tolist()
similars_svd
```

```
similars_svd.insert(0,game)
```

```
original_matrix[original_matrix['Name_Year'].isin(similars_svd)][relavant_cols].sort_values(by=['Name_Year'], key=lambda x: x.map({v: i for i, v in enumerate(similars_svd)}))
```

# Gower Distance

## Original Matrix

```
similars_gower = gower_distances_original.loc[game].tolist()
similars_gower
```

```
similars_gower.insert(0,game)
```

```
original_matrix[original_matrix['Name_Year'].isin(similars_gower)][relavant_cols].sort_values(by=['Name_Year'], key=lambda x: x.map({v: i for i, v in enumerate(similars_gower)}))
```

## Game Theme Mechs

```
similars_gower_reduced = gower_distances_reduced.loc[game].tolist()
similars_gower_reduced
```

```
similars_gower_reduced.insert(0,game)
```

```
original_matrix[original_matrix['Name_Year'].isin(similars_gower_reduced)][relavant_cols].sort_values(by=['Name_Year'], key=lambda x: x.map({v: i for i, v in enumerate(similars_gower_reduced)}))
```
